# MadGraph Project

MadGraph was used to simulate 100 eetautau final state events, from a e plus e minus collider. 

----------------
Syntax for generating the events in Madgraph: 

generate e e~ > e e~ ta ta~

output ee_eetautau

launch ee_eetautau

-----------------


```python
import pylhe
import math
from pprint import *
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
from uproot_methods import TLorentzVector
matplotlib.rcParams['figure.dpi']= 150
```

```python
def calc_pt(px,py):
    '''Calculates transverse momentum'''
    return math.sqrt(px**2+py**2)

def my_hist_plotter(data,x_min,x_max,
                    n_bins,x_axis_name,save,save_name):
    '''Makes basic histogram of data with range specified'''
    fig= plt.figure()
    plt.hist(data,bins=n_bins,range=[x_min, x_max],
             histtype='step',color='tab:blue',fill = True,
             linewidth=0,alpha=0.2, normed = False)
    plt.xlabel(x_axis_name)
    plt.ylabel("Events")
    if save:
        plt.savefig(save_name+".pdf",transparent=True)
    
def my_hist_plotter2(data,n_bins,x_axis_name,save=False,save_name=None,my_range = None):
    '''Makes basic histogram of data without range specified'''
    fig= plt.figure()
    plt.hist(data,bins=n_bins,histtype='step',range=my_range,
             color='tab:blue',fill = True,
             linewidth=0,alpha=0.2, normed = False)
    plt.xlabel(x_axis_name)
    plt.ylabel("Events")

    if save:
        plt.savefig(save_name+".pdf",transparent=True)

#name='/Users/jannicke/Downloads/unweighted_events.lhe'
#name='Events/run_02/unweighted_events.lhe'
name='ee_tautauee/Events/run_03/unweighted_events.lhe'
```

```python
# What sort of things to look at? 
# pT
# phi
# eta 
# mass
# for all particles 

# delta phi between e's and taus
# total energy of the two particles
```

```python
e_plus_pt  = []
e_plus_eta = []
e_plus_phi = []
e_plus_m   = []

e_minus_pt  = []
e_minus_eta = []
e_minus_phi = []
e_minus_m   = []

tau_plus_pt  = []
tau_plus_eta = []
tau_plus_phi = []
tau_plus_m   = []

tau_minus_pt  = []
tau_minus_eta = []
tau_minus_phi = []
tau_minus_m   = []

tau_inv_mass = []
tau_sum_pt = []


# Basic event loop 
for event in pylhe.readLHE(name):
    #e = pd.DataFrame(event['particles'])
    for particle in event['particles']:
        if particle['status'] == 1.0: # final state particle
            px = particle['px']
            py = particle['py']
            pz = particle['pz']
            m = particle['m']
            p = TLorentzVector.from_xyzm( px, py, pz, m)
            if particle['id'] == 11.0: # electron
                e_minus_pt += [p.pt]
                e_minus_phi += [p.phi]
                e_minus_eta += [p.eta]
                e_minus_m += [m]
            if particle['id'] == -11.0: # positive
                e_plus_pt += [p.pt]
                e_plus_phi += [p.phi]
                e_plus_eta += [p.eta]
                e_plus_m += [m]
            if particle['id'] == 15.0: # tau- 
                tau_plus_pt += [p.pt]
                tau_plus_phi += [p.phi]
                tau_plus_eta += [p.eta]
                tau_plus_m += [m]
                tau_plus = p
            if particle['id'] == -15.0: # tau+
                tau_minus_pt += [p.pt]
                tau_minus_phi += [p.phi]
                tau_minus_eta += [p.eta]
                tau_minus_m += [m]
                tau_minus = p
    combined_taus = tau_plus+tau_minus
    tau_inv_mass += [combined_taus.mass]
    tau_sum_pt  += [combined_taus.pt]
            
            
                
```

```python
n_bins = 50 
my_hist_plotter2(e_minus_pt,n_bins,r"$e_-$ $p_T$ [GeV]")
my_hist_plotter2(e_minus_eta,n_bins,r"$e_-$ $\eta$")
my_hist_plotter2(e_minus_phi,n_bins,r"$e_-$ $\phi$")
my_hist_plotter2(e_minus_m,n_bins,r"$e_-$ $m$")

my_hist_plotter2(e_plus_pt,n_bins,r"$e_+$ $p_T$ [GeV]")
my_hist_plotter2(e_plus_eta,n_bins,r"$e_+$ $\eta$")
my_hist_plotter2(e_plus_phi,n_bins,r"$e_+$ $\phi$")
my_hist_plotter2(e_plus_m,n_bins,r"$e_+$ $m$")

my_hist_plotter2(tau_minus_pt,n_bins,r"$\tau_-$ $p_T$ [GeV]")
my_hist_plotter2(tau_minus_eta,n_bins,r"$\tau_-$ $\eta$")
my_hist_plotter2(tau_minus_phi,n_bins,r"$\tau_-$ $\phi$")
my_hist_plotter2(tau_minus_m,n_bins,r"$\tau_-$ $m$")

my_hist_plotter2(tau_plus_pt,n_bins,r"$\tau_+$ $p_T$ [GeV]")
my_hist_plotter2(tau_plus_eta,n_bins,r"$\tau_+$ $\eta$")
my_hist_plotter2(tau_plus_phi,n_bins,r"$\tau_+$ $\phi$")
my_hist_plotter2(tau_plus_m,n_bins,r"$\tau_+$ $m$")
```

```python
n_bins = 40
taus_pt =  tau_plus_pt+tau_minus_pt
taus_eta = tau_plus_eta+tau_minus_eta
taus_phi = tau_plus_phi+tau_minus_phi
my_hist_plotter2(taus_pt,n_bins,r"$\tau$ $p_T$ [GeV]")
my_hist_plotter2(taus_eta,n_bins,r"$\tau$ $\eta$")
my_hist_plotter2(taus_phi,n_bins,r"$\tau$ $\phi$")


es_pt =  e_plus_pt+e_minus_pt
es_eta = e_plus_eta+e_minus_eta
es_phi = e_plus_phi+e_minus_phi
my_hist_plotter2(es_pt,n_bins,r"$e$ $p_T$ [GeV]")
my_hist_plotter2(es_eta,n_bins,r"$e$ $\eta$")
my_hist_plotter2(es_phi,n_bins,r"$e$ $\phi$")
```

```python
# 2D plots 
from matplotlib.colors import LogNorm
def make_2d_hist(x, y, x_label="", y_label="",):
    plt.figure()
    plt.hist2d(x,y,
               range=[[3,5],[0,5]],
               bins=10, 
               norm=LogNorm(),
               cmap='viridis')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.colorbar(label="Events")
    #plt.title("Linear fit on training set")
    plt.legend(loc="upper left",frameon=False)
    return
```
